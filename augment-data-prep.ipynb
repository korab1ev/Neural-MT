```python
import pandas as pd

FILE_PATH = "RU.txt"

df = pd.read_csv(FILE_PATH, sep='\t', header=None, encoding='utf-8')

df.columns = ['geonameid', 'name', 'asciiname', 'alternatenames', 'latitude', 'longitude', 'feature_class', 'feature_code', 'country_code', 'cc2', 'admin1_code', 'admin2_code', 'admin3_code', 'admin4_code', 'population', 'elevation', 'dem', 'timezone', 'modification_date']

df.head(25)
```

```python
#df.drop_duplicates(inplace=True) # row 23 - Yasnaya Polyana, row 24 - Yasnaya Polyana

df = df.drop_duplicates('asciiname', keep='last')
```

```python
import re

ru_en_names_mapping = dict()
en_names_no_russian = []

with open('ru-en-names-augment.txt', mode='w', encoding='utf-8') as f:
    for idx, row in df_with_alternatenames.iterrows():
        #print(row['asciiname'], row['alternatenames'])
        names_list = row['alternatenames'].split(',')
        pr = 0
        for name in names_list:
            if re.search(r'[а-яА-Я]', name):
                pr = 1
                #ru_en_names_mapping[name] = row['asciiname']
                f.write(f"{name}\t{row['asciiname']}\n")
                break
        if pr == 0:
            en_names_no_russian.append(row['asciiname'])
```
